# 3. Data Acquisition and Log Simulation

## 3.2 Data Acquisition and Preprocessing

Generate synthetic log data that mimics real system behavior.

```python
# Define log schema
log_schema = StructType([
    StructField("timestamp", StringType(), True),
    StructField("service", StringType(), True),
    StructField("host", StringType(), True),
    StructField("log_level", StringType(), True),
    StructField("message", StringType(), True),
    StructField("request_id", StringType(), True),
    StructField("response_time_ms", DoubleType(), True),
    StructField("status_code", LongType(), True),
])

print("Log Schema Definition:")
for field in log_schema.fields:
    print(f"  • {field.name:<20} : {field.dataType}")
```

```python
# Synthetic log generator
class SyntheticLogGenerator:
    def __init__(self, services, hosts, anomaly_percentage=0.15, seed=42):
        self.services = services
        self.hosts = hosts
        self.anomaly_percentage = anomaly_percentage
        self.faker = Faker()
        random.seed(seed)
        Faker.seed(seed)
        
        self.SUCCESS_CODES = [200, 201, 204]
        self.ERROR_CODES = [400, 401, 403, 404, 500, 502, 503, 504]
        self.LOG_LEVELS = ['DEBUG', 'INFO', 'WARNING', 'ERROR', 'CRITICAL']
        
        self.ERROR_MESSAGES = [
            "Database connection timeout",
            "Authentication failed",
            "Resource not found",
            "Internal server error",
            "Service unavailable",
            "Request timeout"
        ]
        
        self.NORMAL_MESSAGES = [
            "Request processed successfully",
            "User logged in",
            "Data retrieved",
            "Record created",
            "Cache hit",
            "Health check passed"
        ]
    
    def generate_normal_log(self, base_time):
        return {
            'timestamp': base_time.isoformat(),
            'service': random.choice(self.services),
            'host': random.choice(self.hosts),
            'log_level': random.choices(['DEBUG', 'INFO', 'WARNING'], weights=[0.3, 0.6, 0.1])[0],
            'message': random.choice(self.NORMAL_MESSAGES),
            'request_id': self.faker.uuid4(),
            'response_time_ms': random.randint(50, 500),
            'status_code': random.choice(self.SUCCESS_CODES)
        }
    
    def generate_error_log(self, base_time):
        error_type = random.choices(
            ['high_latency', 'error', 'critical'],
            weights=[0.4, 0.4, 0.2]
        )[0]
        
        if error_type == 'high_latency':
            return {
                'timestamp': base_time.isoformat(),
                'service': random.choice(self.services),
                'host': random.choice(self.hosts),
                'log_level': 'WARNING',
                'message': "High response time detected",
                'request_id': self.faker.uuid4(),
                'response_time_ms': random.randint(3000, 10000),
                'status_code': 200
            }
        else:
            return {
                'timestamp': base_time.isoformat(),
                'service': random.choice(self.services),
                'host': random.choice(self.hosts),
                'log_level': 'ERROR' if error_type == 'error' else 'CRITICAL',
                'message': random.choice(self.ERROR_MESSAGES),
                'request_id': self.faker.uuid4(),
                'response_time_ms': random.randint(100, 1000),
                'status_code': random.choice(self.ERROR_CODES)
            }
    
    def generate_record(self, base_time):
        if random.random() < self.anomaly_percentage:
            return self.generate_error_log(base_time)
        else:
            return self.generate_normal_log(base_time)
    
    def generate_batch(self, num_records, start_time=None):
        if start_time is None:
            start_time = datetime.utcnow()
        
        records = []
        for i in range(num_records):
            time_offset = (i / num_records) * 60  # Spread over 1 minute
            record_time = start_time + timedelta(seconds=time_offset)
            records.append(self.generate_record(record_time))
        
        return records

print("✓ Log Generator defined")
```

```python
# Initialize generator and create sample data
services = ["auth-service", "api-gateway", "payment-service", "user-service", "notification-service"]
hosts = ["host-01", "host-02", "host-03", "host-04", "host-05"]

generator = SyntheticLogGenerator(
    services=services,
    hosts=hosts,
    anomaly_percentage=0.15,
    seed=42
)

# Generate sample logs
num_records = 10000
sample_logs = generator.generate_batch(num_records)

print(f"✓ Generated {num_records} sample log records")
print(f"\nSample Log Records:")
for i in range(3):
    print(f"\nRecord {i+1}:")
    for key, value in sample_logs[i].items():
        print(f"  {key:<15}: {value}")
```

```python
# Create Spark DataFrame from logs via a temporary JSONL file to avoid Python pickling issues
import json
from pathlib import Path

normalized_logs = [
    {
        **record,
        "response_time_ms": int(round(record["response_time_ms"])),
        "status_code": int(record["status_code"]),
    }
    for record in sample_logs
]

base_dir = Path.cwd()
if not (base_dir / "data").exists() and (base_dir.parent / "data").exists():
    base_dir = base_dir.parent

jsonl_path = base_dir / "data" / "sample" / "analysis_sample_logs.jsonl"
jsonl_path.parent.mkdir(parents=True, exist_ok=True)
with jsonl_path.open(mode="w", encoding="utf-8") as temp_file:
    for record in normalized_logs:
        temp_file.write(json.dumps(record) + "\n")

df_logs = spark.read.schema(log_schema).json(str(jsonl_path)).cache()

# Convert timestamp string to timestamp type
df_logs = df_logs.withColumn("timestamp", to_timestamp(col("timestamp")))

print(f"✓ Created Spark DataFrame with {df_logs.count()} records")
print("\nDataFrame Schema:")
df_logs.printSchema()

print("\nFirst 5 Records:")
df_logs.show(5, truncate=False)

# Keep the JSONL file on disk so future notebook actions can reuse it safely
```

# 4. Log Cleaning and Feature Extraction

## 3.2 (continued) Data Preprocessing

Parse raw logs, handle anomalies, and create derived features.

```python
# Data quality checks
def check_data_quality(df):
    total_records = df.count()
    
    # Null value checks
    null_counts = {}
    for column in df.columns:
        null_counts[column] = df.filter(col(column).isNull()).count()
    
    # Response time statistics
    stats = df.select(
        avg(col("response_time_ms")).alias("avg_response_time"),
        spark_max(col("response_time_ms")).alias("max_response_time"),
        spark_min(col("response_time_ms")).alias("min_response_time")
    ).collect()[0]
    
    print("\nData Quality Report:")
    print(f"Total Records: {total_records}")
    print(f"\nNull Value Counts:")
    for col_name, count in null_counts.items():
        print(f"  {col_name}: {count}")
    
    print(f"\nResponse Time Statistics (ms):")
    print(f"  Average: {stats.avg_response_time:.2f}")
    print(f"  Max: {stats.max_response_time:.2f}")
    print(f"  Min: {stats.min_response_time:.2f}")

check_data_quality(df_logs)
```

```python
# Feature engineering
df_processed = df_logs.withColumn(
    "is_error",
    when(
        (col("log_level").isin(["ERROR", "CRITICAL", "FATAL"])) |
        (col("status_code") >= 400),
        1
    ).otherwise(0)
).withColumn(
    "is_high_latency",
    when(col("response_time_ms") > 5000, 1).otherwise(0)
).withColumn(
    "is_server_error",
    when(col("status_code") >= 500, 1).otherwise(0)
).withColumn(
    "is_client_error",
    when((col("status_code") >= 400) & (col("status_code") < 500), 1).otherwise(0)
)

print("✓ Added feature columns:")
print("  • is_error: Log level ERROR/CRITICAL or status >= 400")
print("  • is_high_latency: Response time > 5000ms")
print("  • is_server_error: Status code >= 500")
print("  • is_client_error: Status code 400-499")

print("\nFeature Summary:")
feature_summary = df_processed.select(
    count(when(col("is_error") == 1, 1)).alias("total_errors"),
    count(when(col("is_high_latency") == 1, 1)).alias("high_latency_count"),
    count(when(col("is_server_error") == 1, 1)).alias("server_errors"),
    count(when(col("is_client_error") == 1, 1)).alias("client_errors")
).collect()[0]

print(f"Total Errors: {feature_summary.total_errors}")
print(f"High Latency Events: {feature_summary.high_latency_count}")
print(f"Server Errors: {feature_summary.server_errors}")
print(f"Client Errors: {feature_summary.client_errors}")
```

# 6. Real-Time Filtering, Counting, and Anomaly Detection

## 3.5 Application of Design Patterns

```python
# DESIGN PATTERN 1: FILTERING
print("\n" + "="*70)
print("DESIGN PATTERN 1: FILTERING")
print("="*70)

# Filter only error events
error_keywords = ["error", "failure", "failed", "exception", "critical"]
error_df = df_processed.filter(
    (col("is_error") == 1) |
    col("message").rlike(f"(?i){'|'.join(error_keywords)}")
)

error_count = error_df.count()
total_count = df_processed.count()
filtering_reduction = ((total_count - error_count) / total_count) * 100

print(f"\nFilter Logic: Keep only ERROR, CRITICAL events + status >= 400")
print(f"Input Records: {total_count}")
print(f"Output Records (Errors): {error_count}")
print(f"Reduction: {filtering_reduction:.1f}% (efficiency improvement)")

print("\nError Distribution by Log Level:")
error_by_level = error_df.groupBy("log_level").count().collect()
for row in sorted(error_by_level, key=lambda x: x[1], reverse=True):
    percentage = (row[1] / error_count) * 100
    print(f"  {row[0]:<10}: {row[1]:>5} ({percentage:>5.1f}%)")
```

```python
# DESIGN PATTERN 2: AGGREGATION
print("\n" + "="*70)
print("DESIGN PATTERN 2: AGGREGATION (By Service)")
print("="*70)

from pyspark.sql.functions import count as spark_count

# Aggregate by service
service_agg = error_df.groupBy("service").agg(
    spark_count(lit(1)).alias("event_count"),
    avg(col("response_time_ms")).alias("avg_response_time"),
    spark_max(col("response_time_ms")).alias("max_response_time"),
    spark_count(when(col("status_code") >= 500, 1)).alias("server_errors"),
    spark_count(when(col("status_code") >= 400, 1)).alias("client_errors")
)

service_totals = df_processed.groupBy("service").agg(spark_count(lit(1)).alias("total_events"))
service_agg_pd = service_agg.join(service_totals, on="service", how="left").toPandas()
service_agg_pd["error_rate"] = (service_agg_pd["server_errors"] + service_agg_pd["client_errors"]) / service_agg_pd["total_events"]
service_agg_pd = service_agg_pd.sort_values("event_count", ascending=False)

print("\nService-Level Aggregation (Errors Only):")
print(service_agg_pd.to_string(index=False))
```

```python
# DESIGN PATTERN 3: SORTING
print("\n" + "="*70)
print("DESIGN PATTERN 3: SORTING (Top N Analysis)")
print("="*70)

top_failing_services = service_agg_pd.nlargest(3, 'event_count')[['service', 'event_count', 'avg_response_time', 'error_rate']]

print("\nTop 3 Failing Services:")
for idx, row in top_failing_services.iterrows():
    print(f"\n  {row['service']}:")
    print(f"    Error Events: {int(row['event_count'])}")
    print(f"    Avg Response Time: {row['avg_response_time']:.2f}ms")
    print(f"    Error Rate: {row['error_rate']:.1%}")
```

```python
# DESIGN PATTERN 4: WINDOWING & AGGREGATION
print("\n" + "="*70)
print("DESIGN PATTERN 4: WINDOWING (Time-Based Aggregation)")
print("="*70)

# Create 60-second windows with 10-second slides
windowed_agg = (
    error_df.groupBy(
        window(col("timestamp"), "60 seconds", "10 seconds"),
        col("service")
    )
    .agg(
        count(lit(1)).alias("event_count"),
        avg(col("response_time_ms")).alias("avg_response_time"),
        count(when(col("status_code") >= 500, 1)).alias("server_errors")
    )
    .withColumn("error_rate", spark_round(col("server_errors") / col("event_count"), 4))
    .withColumn("window_start", col("window.start"))
    .withColumn("window_end", col("window.end"))
)

windowed_pd = windowed_agg.select("window_start", "window_end", "service", "event_count", "avg_response_time", "error_rate").toPandas()

print("\nWindowed Aggregation (60-second windows, 10-second slides):")
print(windowed_pd[['window_start', 'service', 'event_count', 'avg_response_time', 'error_rate']].head(10).to_string(index=False))
print(f"\nTotal windows: {windowed_pd['window_start'].nunique()}")
```

```python
# ANOMALY DETECTION
print("\n" + "="*70)
print("ANOMALY DETECTION ENGINE")
print("="*70)

# Thresholds for anomaly detection
RESPONSE_TIME_THRESHOLD = 5000  # ms
ERROR_RATE_THRESHOLD = 0.15  # 15%
SERVER_ERROR_THRESHOLD = 0.10  # 10%

# Add anomaly flags
anomalies = windowed_agg.withColumn(
    "high_latency",
    when(col("avg_response_time") > lit(RESPONSE_TIME_THRESHOLD), True).otherwise(False)
).withColumn(
    "high_error_rate",
    when(col("error_rate") > lit(ERROR_RATE_THRESHOLD), True).otherwise(False)
).withColumn(
    "is_anomaly",
    when(col("high_latency") | col("high_error_rate"), True).otherwise(False)
).filter(
    col("is_anomaly") == True
)

anomaly_pd = anomalies.toPandas()

print(f"\nAnomalies Detected: {len(anomaly_pd)}")
print(f"\nThresholds:")
print(f"  • Response Time: > {RESPONSE_TIME_THRESHOLD}ms")
print(f"  • Error Rate: > {ERROR_RATE_THRESHOLD:.1%}")

if len(anomaly_pd) > 0:
    print(f"\nTop Anomalies:")
    anomaly_display = anomaly_pd[['window_start', 'service', 'event_count', 'avg_response_time', 'error_rate']].head(5)
    for idx, row in anomaly_display.iterrows():
        print(f"\n  {row['service']} @ {row['window_start']}")
        print(f"    Events: {int(row['event_count'])}, Avg Latency: {row['avg_response_time']:.0f}ms, Error Rate: {row['error_rate']:.1%}")
else:
    print("\n  No anomalies detected in this batch")
```
